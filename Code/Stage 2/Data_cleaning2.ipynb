```python
#Reads in regular season results (train), tournament results (test)
#and kenpom data (supplemental)

kenpom_data=pd.read_csv(filepath1+'kenpom_data_stage2.csv')
reg_season=pd.read_csv(filepath+'regular_season_compact_results.csv')
tourney=pd.read_csv(filepath+'tourney_compact_results.csv')
teams=pd.read_csv(filepath+'teams.csv')

reg_season_15=pd.read_csv(filepath+'regular_season_compact_results_2015.csv')
```

```python
reg_season = reg_season.append(reg_season_15)
```

```python
#Only considers regular season games played after 2002
#and the 2011-2014 NCAA Tournaments:

reg_season=reg_season[reg_season.season >= 2003]
tourney=tourney[tourney.season >= 2003]
```

```python
#Ratio of tournament games to season games
tourney_ratio=float(len(tourney))/len(reg_season)
```

```python
#Indicates if game is tournament game
reg_season['tourney_game']=np.array([0.0]*len(reg_season))
tourney['tourney_game']=np.array([1.0]*len(tourney))
```

```python
#Gets training data - regular season and tournament

training_data=reg_season.append(tourney)
```

```python
#Join kenpom data and teams

kenpom_data=kenpom_data.merge(teams,how='left',on='team_name')
```

```python
#Adjust for homecourt advantage:
homecourt=pd.read_csv(filepath+'homecourt.csv')
def get_adj_gamescore_winner(x,homecourt): 
    points = x.wscore
    if x.wloc == 'N':
        return points
    if x.wloc == 'H':
        team=x.wteam
        point_diff=float(homecourt[homecourt.team_id == team]['point_diff'])
        return points-point_diff/2.
    if x.wloc == 'A':
        team=x.lteam
        point_diff=float(homecourt[homecourt.team_id == team]['point_diff'])
        return points+point_diff/2.
    
    
def get_adj_gamescore_loser(x,homecourt):
    points = x.lscore
    if x.wloc == 'N':
        return points
    if x.wloc == 'H':
        team=x.wteam
        point_diff=float(homecourt[homecourt.team_id == team]['point_diff'])
        return points+point_diff/2.
    if x.wloc == 'A':
        team=x.wteam
        point_diff=float(homecourt[homecourt.team_id == team]['point_diff'])
        return points-point_diff/2.
        
training_data['wscore']=training_data.apply(lambda x: get_adj_gamescore_winner(x,homecourt),axis=1)
training_data['lscore']=training_data.apply(lambda x: get_adj_gamescore_loser(x,homecourt),axis=1)
```

```python
#Gets the pythagorean win percentage of the winning team 
#and associated weight:

def pythagorean(w_score,l_score):
    return w_score**13.91/(w_score**13.91+l_score**13.91)

def get_pythag(x):
    if x.numot == 0:
        return pythagorean(x.wscore,x.lscore)
    else:
        return .50
    
def get_gamescore_weight(x):
    pyth = get_pythag(x)
    return 1/(1-(pyth-.5))


#I want to make regular season and tournament games weighted 50/50,
#so I use the tourney_ratio to weigh tourney games

def get_tourney_weight(x,ratio):
    if x.tourney_game==1:
        return 1./ratio
    if x.tourney_game==0:
        return 1
        


#Gets weight associated with the day the game was played on:
def get_day_weight(x):
    day=x.daynum
    return (float(day)/132)**4/2+1

#Gets total weight based on day weight and game score weight:
def get_total_weight(x,tourney_ratio):
    day_weight=get_day_weight(x)
    score_weight=get_gamescore_weight(x)
    tourney_weight=get_tourney_weight(x,tourney_ratio)
    
    return day_weight*score_weight*tourney_weight
    
    
training_data['game_weight']=training_data.apply(lambda x: get_total_weight(x,tourney_ratio),axis=1)
```

```python
#Randomly assign teams as team_1 and team_2, and set
#result=1 if team_1 was the winner
import random
def assign_teams(df):
    team_1=[]
    team_2=[]
    result=[]
    for ind in df.index.values:
        d=df.ix[ind]
        a=np.random.random()
        if a < .5:
            team_1.append(d.wteam)
            team_2.append(d.lteam)
            result.append(1)
        else:
            team_1.append(d.lteam)
            team_2.append(d.wteam)
            result.append(0)
    df['team_1'] = np.array(team_1)
    df['team_2'] = np.array(team_2)
    df['result'] = np.array(result)
```

```python
training_data=training_data.reset_index(drop='index')
```

```python
assign_teams(training_data)
training_data = training_data.drop(['daynum','wteam','wscore','lteam','lscore','numot','tourney_game'],axis=1)
```

```python
###Get testing data

tourney_teams=pd.read_csv(filepath+'tourney_seeds_2015.csv')
tourney_teams=tourney_teams.merge(teams,how='left',on='team_id')
```

```python
team_1=[]
team_2=[]
team_1_id=[]
team_2_id=[]
for ind in tourney_teams.index.values:
    for i in range(ind+1,len(tourney_teams)):
        team_1.append(tourney_teams.ix[ind].team_name)
        team_2.append(tourney_teams.ix[i].team_name)
        team_1_id.append(tourney_teams.ix[ind].team_id)
        team_2_id.append(tourney_teams.ix[i].team_id)
```

```python
testing_data=pd.DataFrame({'team_1':np.array(team_1),'team_2':np.array(team_2),
                           'team_1_id':np.array(team_1_id),'team_2_id':np.array(team_2_id)})
testing_data['season']=np.array([2015]*len(testing_data))
```

```python
#Gets features from kenpom_data
kenpom_data.columns.tolist()
not_features=['season','team_name','team_id','year']
features=[col for col in kenpom_data.columns.tolist() if col not in not_features]
```

```python
def get_feature(x,team_num,feat,data):
    team=x[team_num]
    seas=x.season
    if len(data[data.team_id==team][data.year==seas][feat] > 0):
        return float(data[data.team_id==team][data.year==seas][feat])
    else:
        return np.nan
```

```python
#*** THIS TAKES A LONG TIME

#Get features for team 1 in training data
for feat in features:
    col_name='team_1_'+feat
    training_data[col_name]=training_data.apply(lambda x: get_feature(x,'team_1',feat,kenpom_data),axis=1)
    print col_name
    
#Get features for team 2 in training data
for feat in features:
    col_name='team_2_'+feat
    training_data[col_name]=training_data.apply(lambda x: get_feature(x,'team_2',feat,kenpom_data),axis=1)
    print col_name
```

```python
#***THIS ALSO TAKES A LONG TIME

#Get features for team 1 in test
for feat in features:
    col_name='team_1_'+feat
    testing_data[col_name]=testing_data.apply(lambda x: get_feature(x,'team_1_id',feat,kenpom_data),axis=1)
    print col_name
    
#Get features for team 2 in test
for feat in features:
    col_name='team_2_'+feat
    testing_data[col_name]=testing_data.apply(lambda x: get_feature(x,'team_2_id',feat,kenpom_data),axis=1)
    print col_name
```
